```python
reviews_json_file_path = 'Pet_Supplies.json'
df = pd.read_json(reviews_json_file_path, lines=True)

print("Number of rows in the dataset:", df.shape[0])
print("Number of columns in the dataset:", df.shape[1])
```

#### The first 100 users with the most purchases with the corresponding number of purchases (NumberOfReviews column) and the number of users who purchased the same products (NumberOfUsers column)

```python
reviewer_counts = df['reviewerID'].value_counts()
reviewer_counts_df = reviewer_counts.reset_index()
```

```python
pd.set_option('display.max_rows', None)
```

```python
top_reviewers = reviewer_counts_df.head(100)['reviewerID']
top_reviewer_ids = top_reviewers.tolist()
results = []

for reviewer_id in top_reviewer_ids:
    reviewer_df = df[df['reviewerID'] == reviewer_id]
    num_reviews = len(reviewer_df)

    unique_asin_values = reviewer_df['asin'].unique()
    filtered_df = df[df['asin'].isin(unique_asin_values)]
    num_reviewer_ids = filtered_df['reviewerID'].nunique()
    
    results.append({
        'reviewerID': reviewer_id,
        'NumberOfReviews': num_reviews,
        'NumberOfUsers': num_reviewer_ids
    })

results_df = pd.DataFrame(results)
results_df
```

#### Filtering the purchases of only a specific user

```python
reviewerID = "A22CW0ZHY3NJH8"
specific_reviewer_df = df[df['reviewerID'] == reviewerID]
specific_reviewer_df.head()
```

#### The number of unique products purchased by this single specific

```python
unique_asin_count = specific_reviewer_df['asin'].nunique()
print("Number of unique products:", unique_asin_count)
```

#### The number of users who purchased the same products

```python
unique_asin_values = specific_reviewer_df['asin'].unique()
filtered_df = df[df['asin'].isin(unique_asin_values)]
reviewer_ids = filtered_df['reviewerID'].unique()
reviewer_ids_list = reviewer_ids.tolist()
print("Number of users who purchased the same products:", len(reviewer_ids))
```

#### Iteration through all identified users and for each product that was purchased by both an identified user and our specific user (one of 15 selected), we searched for 2 products (chronologically ordered) that the identified user had purchased before buying the given product.

```python
result_list = []

for reviewer_id in reviewer_ids_list:
    reviewer_df = df[df['reviewerID'] == reviewer_id]

    for product_id in unique_asin_values:
        sorted_products = reviewer_df.sort_values(by='unixReviewTime').reset_index(drop=True)
        product_indices = sorted_products[sorted_products['asin'] == product_id].index

        if not product_indices.empty and product_indices.min() >= 2:
            selected_products = sorted_products.loc[product_indices.min() - 2:product_indices.min() - 1, 'asin'].tolist()
            result_list.append({'reviewerID': reviewer_id, 'productID': product_id, 'selected_products': selected_products})
```

```python
with open('users-2.json', 'w') as json_file:
    json.dump(result_list, json_file)
```
