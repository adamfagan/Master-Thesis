```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
file_path = "/kaggle/input/experiment-2-1-trainings-info/experiment-2_1-trainings-info.json"

with open(file_path, "r") as file:
    users = json.load(file)

df_training_info = pd.DataFrame(users)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_training_info
```

```python
columns_to_plot = [
    'model_lstm_test_f1_score_k_1',
    'model_bi_lstm_test_f1_score_k_1',
    'model_gru_test_f1_score_k_1',
    'model_autoencoder_test_f1_score_k_1'
]

labels = {
    'model_lstm_test_f1_score_k_1': 'LSTM',
    'model_bi_lstm_test_f1_score_k_1': 'Bidirectional LSTM',
    'model_gru_test_f1_score_k_1': 'GRU',
    'model_autoencoder_test_f1_score_k_1': 'Autoencoder'
}

plt.figure(figsize=(14, 7))
for column in columns_to_plot:
    plt.plot(df_training_info['user'], df_training_info[column], marker='o', label=labels[column])

plt.title('Models performance')
plt.xlabel('User')
plt.ylabel('Test weighted F1-score at k=1')
plt.xticks(df_training_info['user'].unique(), rotation=45, ha='right', va='top')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
