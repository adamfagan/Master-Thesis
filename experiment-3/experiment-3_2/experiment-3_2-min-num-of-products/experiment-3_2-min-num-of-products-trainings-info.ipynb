```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
file_path = "/kaggle/input/experiment-3-2-trainings-min-num-of-products-info/experiment-3_2-trainings-min-num-of-products-info.json"

with open(file_path, "r") as file:
    users = json.load(file)

df_training_info = pd.DataFrame(users)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_training_info
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df_training_info['min_num_of_products_in_sequence'], df_training_info['model_bi_lstm_test_f1_score'], 'o-', label='Test weighted F1-score at k=5')
plt.plot(df_training_info['min_num_of_products_in_sequence'], df_training_info['model_bi_lstm_test_hr'], 'o-', label='Test HR at k=5')
plt.plot(df_training_info['min_num_of_products_in_sequence'], df_training_info['model_bi_lstm_test_ndcg'], 'o-', label='Test NDCG at k=5')

plt.xlabel('Minimum number of products purchased')
plt.ylabel('Metrics values')
plt.title('Model performance')
plt.legend()
plt.xticks(df_training_info['min_num_of_products_in_sequence'].unique())
plt.grid(True)
plt.show()
```
