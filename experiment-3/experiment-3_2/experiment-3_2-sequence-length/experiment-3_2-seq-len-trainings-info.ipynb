```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
file_path = "/kaggle/input/experiment-3-2-trainings-seq-len-info/experiment-3_2-trainings-seq-len-info.json"

with open(file_path, "r") as file:
    users = json.load(file)

df_training_info = pd.DataFrame(users)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_training_info
```

```python
columns_to_plot = [
    'model_lstm_test_hr',
    'model_bi_lstm_test_hr',
    'model_gru_test_hr',
    'model_autoencoder_test_hr',
    'model_cnn_test_hr'
]

labels = {
    'model_lstm_test_hr': 'LSTM',
    'model_bi_lstm_test_hr': 'Bidirectional LSTM',
    'model_gru_test_hr': 'GRU',
    'model_autoencoder_test_hr': 'Autoencoder',
    'model_cnn_test_hr': 'CNN'
}

plt.figure(figsize=(10, 6))
for column in columns_to_plot:
    plt.plot(df_training_info['sequence_length'], df_training_info[column], marker='o', label=labels[column])

plt.title('Models performance')
plt.xlabel('Sequence length')
plt.ylabel('Test HR at k=5')
plt.xticks(df_training_info['sequence_length'].unique())
plt.grid(True)
plt.legend()
plt.show()
```
